```python

from matplotlib.pyplot import tight_layout


alphas = [2,3,4,5]

fig,axs = plt.subplots(2,1,figsize=[10/2.54,11/2.54],tight_layout=True)

resultfile4 = f"synthetic_data_S1/Varyp_crust_mantle_gradMoho_synthetics_c35_Vpc6.37_kc1.78_kcl1.78_Tcl5_TlV2_THV2_hs4_Vps3.6_Ks2.4_dt0.01/recovery_results_diffalpha.npz"
data = np.load(resultfile4)
Hcs = data["Hc_recovery"]
Kcs = data["Kc_recovery"]
axs[0].plot(alphas,Hcs,color='k',marker='s',ms=10,lw=2)
axs[1].plot(alphas,Kcs,color='k',marker='s',ms=10,lw=2)


markers= ['o','^','s']
hlcs = [5,7.5,10] 
hlv = 2

for i,hlc in enumerate(hlcs):
    resultfile4 = f"synthetic_data_S1/Varyp_crust_mantle_gradMoho_synthetics_c35_Vpc6.37_kc1.78_kcl1.78_Tcl{hlc}_TlV{hlv}_THV2_hs4_Vps3.6_Ks2.4_dt0.01/recovery_results_diffalpha.npz"
    data = np.load(resultfile4)
    Hcs = data["Hc_recovery"]
    Kcs = data["Kc_recovery"]
    axs[0].plot(alphas,Hcs,color='b',marker=markers[i],lw=0.5,ls='--')
    axs[1].plot(alphas,Kcs,color='b',marker=markers[i],lw=0.5,ls='--')


hlcs = [5,7.5,10] 
hlv = 0 
for i,hlc in enumerate(hlcs):
    resultfile4 = f"synthetic_data_S1/Varyp_crust_mantle_gradMoho_synthetics_c35_Vpc6.37_kc1.78_kcl1.78_Tcl{hlc}_TlV{hlv}_THV2_hs4_Vps3.6_Ks2.4_dt0.01/recovery_results_diffalpha.npz"
    data = np.load(resultfile4)
    Hcs = data["Hc_recovery"]
    Kcs = data["Kc_recovery"]
    axs[0].plot(alphas,Hcs,color='r',marker=markers[i],lw=0.5)
    axs[1].plot(alphas,Kcs,color='r',marker=markers[i],lw=0.5)



axs[0].set_xlabel("Gaussian Param")
axs[0].set_ylabel("Hc [km]")
axs[0].set_ylim(25,40)


axs[1].set_xlabel("Gaussian Param")
axs[1].set_ylabel("Kc")
axs[1].set_ylim(1.5,1.9)

plt.savefig("Figures/Hk_diffalpha.pdf",dpi=900)

plt.show()

# resultfile1 = "synthetic_data_S1/Varyp_crust_mantle_gradMoho_synthetics_c35_Vpc6.37_kc1.75_Vplc7.2_klc1.8_tlc0_hs4_Vps3.6_Ks2.4_dt0.01/sediment_depth4/mod.npz"
# resultfile2 = "synthetic_data_S1/Varyp_crust_mantle_gradMoho_synthetics_c35_Vpc6.37_kc1.75_Vplc7.2_klc1.8_tlc5_hs4_Vps3.6_Ks2.4_dt0.01/sediment_depth4/mod.npz"
# resultfile3 = "synthetic_data_S1/Varyp_crust_mantle_gradMoho_synthetics_c35_Vpc6.37_kc1.75_Vplc7.2_klc1.8_tlc10_hs4_Vps3.6_Ks2.4_dt0.01/sediment_depth4/mod.npz"
# resultfile4 = "synthetic_data_S1/Varyp_crust_mantle_gradMoho_synthetics_c35_Vpc6.37_kc1.78_kcl1.78_Tcl5_TlV0_THV2_hs4_Vps3.6_Ks2.4_dt0.01/sediment_depth4/mod.npz"
# resultfile5 = "synthetic_data_S1/Varyp_crust_mantle_gradMoho_synthetics_c35_Vpc6.37_kc1.78_kcl1.78_Tcl5_TlV2_THV2_hs4_Vps3.6_Ks2.4_dt0.01/sediment_depth4/mod.npz"

# resultfile6 = "synthetic_data_S1/Varyp_crust_mantle_gradMoho_synthetics_c35_Vpc6.37_kc1.78_kcl1.78_Tcl10_TlV0_THV2_hs4_Vps3.6_Ks2.4_dt0.01/sediment_depth4/mod.npz"

# resultfile7 = "synthetic_data_S1/Varyp_crust_mantle_gradMoho_synthetics_c35_Vpc6.37_kc1.78_kcl1.78_Tcl10_TlV2_THV2_hs4_Vps3.6_Ks2.4_dt0.01/sediment_depth4/mod.npz"



```
