```python
from HkVp_multilayer.plotting import Plot_HkVp
import matplotlib.pyplot as plt 
import numpy as np 
k=2.3
depths=[1, 2, 3, 4, 5, 6, 7, 8, 9]
Vpcs_opt = []
for depth in depths:
    save_filename = f"Hk_result_v75_10_vp_Hk_result.npz"
    resultdir = f"synthetic_data_S1.0/Varyp_sediment_crust_mantle_synthetics_c35_ks{k}_dt0.01/sediments_depth{depth}"
    plot_instance = Plot_HkVp(savefile=f"{resultdir}/{save_filename}")
    Vpcs_opt.append(plot_instance.get_HkVp_opt()[0])

fig,ax = plt.subplots(1,1,figsize=(15/2.54,6/2.54))
ax.plot(depths,Vpcs_opt)
ax.set_ylim(6.0,7.5)
plt.show()

```
